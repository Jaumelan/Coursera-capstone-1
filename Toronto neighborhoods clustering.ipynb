```python
source= requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
soup=BeautifulSoup(source,'lxml')
tab=soup.find('table', class_='wikitable sortable').tbody.text
```

```python
second_list=tab.splitlines()
```

```python
i=0 
length = len(second_list)  
while(i<length):
    if(second_list[i]==''):
        second_list.remove (second_list[i])
        length = length -1  
        continue
    i = i+1
```

```python
labels=['PostalCode','Borough','Neighborhood']
```

```python
chunks = [second_list[x:x+3] for x in range(0, len(second_list), 3)]
```

```python
wiki = pd.DataFrame.from_records(chunks, columns=labels)
```

```python
wiki.drop([0],axis=0,inplace=True)
```

```python
wiki.reset_index()
```

```python
wiki.drop(wiki[wiki['Borough']=="Not assigned"].index,axis=0, inplace=True)
```

```python
wiki1=wiki.groupby("PostalCode").agg(lambda x:','.join(set(x)))
```

```python
wiki1.loc[wiki1['Neighborhood']=="Not assigned",'Neighborhood']=wiki1.loc[wiki1['Neighborhood']=="Not assigned",'Borough']
```

```python
wiki2=wiki1.reset_index()
```
